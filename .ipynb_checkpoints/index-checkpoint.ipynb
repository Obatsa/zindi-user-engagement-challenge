# Exploratory Data Analysis

```python
## percentage completeness
def completeness(df):
    """This function checks percentage completeness in the dataframe"""
    columns = list(df.columns)
    length = len(df)
    perc_completeness = []
    
    for column in columns:
        # finding missing values
        missing = df[column].isna().sum()
        # changing to perc
        perc = (1 - (missing / length)) * 100
        # appending
        perc_completeness.append(perc)
    
    # converting to dataframe
    output = pd.DataFrame(list(zip(columns, perc_completeness)), 
                         columns = ['features', '%_completeness'])
    # sort
    output = output.sort_values(by= '%_completeness', ascending= True)
    output.reset_index(inplace= True, drop= True)
    
    return output
```

```python
# function for one hot encoding
def ohe_col(df, column):
    """This function takes in the name of a dataframe and the categorical column name
    and then uses the oneHotEncoder method from Scikit learn to encode the column"""
    
    # Create a variable to hold the encoded
    column_vals = df[[column]]
    
    # Instatiating a oneHotEcoder with categories = "auto", sparse = False and handle_unknown = "ignore"
    ohe = OneHotEncoder(sparse = False, handle_unknown = "ignore")
    
    # fitting the ohe on the column_vals
    ohe.fit(column_vals)
    
    # Inspecting the categories of the fitted encoder
    print("The categories fitted are", ohe.categories_)
    
    # Transforming the column categories using the ohe and assigning the variable to a variable
    column_vals_encoded = ohe.transform(column_vals)
    
    # Print the transformed values
    print("Transformed data is: ", column_vals_encoded)
    
    # Changing the transformed data into a dataframe
    column_vals_encoded = pd.DataFrame(
        # Parsing in the numpy array
        column_vals_encoded,
        # Setting column names as categories found by OHE
        columns = ohe.categories_[0],
        # Setting index to match the df index
        index=df.index
    )
    
    return column_vals_encoded
```

From the above features columns, most of the columns are lists. The lists will be split during the preprocessing phase while simultaneously one hot encoding.

```python
# dont forget to look for splitting solutions
```

Below two functions are created to split the different lists in the features column, create a list of features and finally create a dataframe of one hot encoded columns.

```python
def feature_list(df, column):
    """This function takes in a dataframe and a column and returns unique values
    found in the column"""
    feature = []
    
    # finding and creating the the list of all elements
    for i in range(len(df)):
        feat_list = df.loc[i, column]
        
        for j in feat_list:
            for k in range(j+1):
                if k not in feature:
                    feature.append(k)
    
    # Returning the feature list
    return feature
```

```python
# Find values in list & Create new values
def create_df(df, column):
    """This function takes in a dataframe and create new columns from the unique values,
    Renames the columns using the unique features, then one hot encodes the columns"""
    features = feature_list(df, column)
    print('This are the unique features', features)
    # renaming features
    columns = [f'{column}_'+'{}'.format(x) for x in features]
    
    # creating zeroes in df
    for f in features:
        df[f] = np.zeros(len(df))
    
    # filling the new columns with 1 and 0 if none
    for i in df.index:
        records = df.loc[i, column]
        
        for f in records:
            df.loc[i, f] = 1
    
    # rename the columns
    rename_col = {features[i]: columns[i] for i in range(len(features))}
    df.rename(columns = rename_col, inplace= True)
    # Check later
    
    # dropping columns
    df.drop([column], axis= 1, inplace= True)
    # return df
    return df
```

## Modeling

```python
# https://www.quora.com/I-have-unlabeled-data-without-a-target-variable-How-do-I-create-a-target-variable-I-have-to-classify-the-data-into-2-groups-ultimately
# https://h2o.ai/wiki/target-variable/
# https://pandas.pydata.org/docs/dev/user_guide/merging.html
# https://sparkbyexamples.com/pandas/pandas-merge-two-dataframes-on-multiple-columns/
# https://www.ablebits.com/office-addins-blog/excel-vlookup-sum-sumif/
```

Proceeding with the two combined dataframes. The models tested will be:
1. Logistic Regression model
2. KNN model
3. Decision tree model
4. Random forest model
5. Neural Networks model

```python
def model_preds(model, x_train, x_test, y_train, y_test):
    """This function takes in the model, train and test data then makes
    prediction then prints the classification report"""
    
    # fitting model
    model_ = model.fit(x_train, y_train)
    
    # making preds
    y_train_preds = model_.predict(x_train)
    y_test_preds = model_.predict(x_test)
    
    # printing MAE
    print('Train MAE: ', round(mean_absolute_error(y_train, y_train_preds), 4))
    print('Train MAE: ', round(mean_absolute_error(y_test, y_test_preds), 4))
    print('*'* 100)
    print('Classification training report')
    print(classification_report(y_train, y_train_preds))
    print('*'* 100)
    print('Classification test report')
    print(classification_report(y_test, y_test_preds))
    print('*'* 100)
    print('Confusion matrix for the training set')
    plot_confusion_matrix(model, x_train, y_train, normalize= 'true');
    print('*'* 100)
    print('Confusion matrix for the testing set')
    plot_confusion_matrix(model, x_test, y_test, normalize= 'true');
```
